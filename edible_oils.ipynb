```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import opendatasets as od
```

```python
oildata = od.download('https://www.kaggle.com/datasets/chinmayshanbhag/historical-price-data-of-8-edible-oils-commodity')
```

```python
dataset = pd.read_csv('/content/historical-price-data-of-8-edible-oils-commodity/8 Edible Oils.csv')
```

```python
dataset.head()
```

```python
dataset.dtypes
```

```python
dataset['Month'] = pd.to_datetime(dataset['Month'])
```

```python
dataset.dtypes
```

```python
dataset.head()
```

```python
 dataset.head(n = 25) 
```

```python
dataset = dataset.loc[dataset['Month'].dt.year >= 2002, :]
```

```python
dataset.head(n = 15)
```

```python
dataset = dataset.dropna()

dataset.head(n = 20)
```

```python
dataset = dataset.drop(dataset.index[0])
dataset.head()
```

```python
dataset = dataset.rename(columns = {
    'Month': 'month',
    'Coconut Oil Price': 'coconut_oil_price',
    'Coconut Oil Price Change': 'coconut_oil_change',
    'Olive Oil Price': 'olive_oil_price',
    'Olive Oil Price Change': 'olive_oil_change',
    'Palm Kernel Oil Price': 'palmkernel_oil_price',
    'Palm Kernel Oil Price Change': 'palmkernel_oil_change',
    'Palm Oil Price': 'palm_oil_price',
    'Palm Oil Price Change': 'palm_oil_change',
    'Peanut Oil Price': 'peanut_oil_price',
    'Peanut Oil Price Change': 'peanut_oil_change',
    'Rapeseed Oil Price': 'rapeseed_oil_price',
    'Rapeseed Oil Price Change': 'rapeseed_oil_change',
    'Soybean Oil Price': 'soybean_oil_price',
    'Soybean Oil Price Change': 'soybean_oil_change',
    'Sunflower Oil Price': 'sunflower_oil_price',
    'Sunflower Oil Price Change': 'sunflower_oil_change'
})
```

```python
dataset.head()
```

```python
dataset = dataset.convert_dtypes(convert_string = False, convert_boolean = False)
```

```python
dataset.dtypes
```

```python
dataset.columns.tolist()
```

```python
dataset['coconut_oil_price'] = dataset['coconut_oil_price'].str.replace(',', '')
dataset['coconut_oil_change'] = dataset['coconut_oil_change'].str.replace(',', '')
dataset['olive_oil_price'] = dataset['olive_oil_price'].str.replace(',', '')
dataset['olive_oil_change'] = dataset['olive_oil_change'].str.replace(',', '')
dataset['palmkernel_oil_price'] = dataset['palmkernel_oil_price'].str.replace(',', '')
dataset['coconut_oil_price'] = dataset['coconut_oil_price'].str.replace(',', '')
dataset['palmkernel_oil_change'] = dataset['palmkernel_oil_change'].str.replace(',', '')
dataset['palm_oil_price'] = dataset['palm_oil_price'].str.replace(',', '')
dataset['palm_oil_change'] = dataset['palm_oil_change'].str.replace(',', '')
dataset['peanut_oil_price'] = dataset['peanut_oil_price'].str.replace(',', '')
dataset['peanut_oil_change'] = dataset['peanut_oil_change'].str.replace(',', '')
dataset['rapeseed_oil_price'] = dataset['rapeseed_oil_price'].str.replace(',', '')
dataset['rapeseed_oil_change'] = dataset['rapeseed_oil_change'].str.replace(',', '')
dataset['soybean_oil_price'] = dataset['soybean_oil_price'].str.replace(',', '')
dataset['soybean_oil_change'] = dataset['soybean_oil_change'].str.replace(',', '')
dataset['sunflower_oil_price'] = dataset['sunflower_oil_price'].str.replace(',', '')
dataset['sunflower_oil_change'] = dataset['sunflower_oil_change'].str.replace(',', '')
```

```python
dataset['coconut_oil_price'] = dataset['coconut_oil_price'].str.replace('%', '')
dataset['coconut_oil_change'] = dataset['coconut_oil_change'].str.replace('%', '')
dataset['olive_oil_price'] = dataset['olive_oil_price'].str.replace('%', '')
dataset['olive_oil_change'] = dataset['olive_oil_change'].str.replace('%', '')
dataset['palmkernel_oil_price'] = dataset['palmkernel_oil_price'].str.replace('%', '')
dataset['coconut_oil_price'] = dataset['coconut_oil_price'].str.replace('%', '')
dataset['palmkernel_oil_change'] = dataset['palmkernel_oil_change'].str.replace('%', '')
dataset['palm_oil_price'] = dataset['palm_oil_price'].str.replace('%', '')
dataset['palm_oil_change'] = dataset['palm_oil_change'].str.replace('%', '')
dataset['peanut_oil_price'] = dataset['peanut_oil_price'].str.replace('%', '')
dataset['peanut_oil_change'] = dataset['peanut_oil_change'].str.replace('%', '')
dataset['rapeseed_oil_price'] = dataset['rapeseed_oil_price'].str.replace('%', '')
dataset['rapeseed_oil_change'] = dataset['rapeseed_oil_change'].str.replace('%', '')
dataset['soybean_oil_price'] = dataset['soybean_oil_price'].str.replace('%', '')
dataset['soybean_oil_change'] = dataset['soybean_oil_change'].str.replace('%', '')
dataset['sunflower_oil_price'] = dataset['sunflower_oil_price'].str.replace('%', '')
dataset['sunflower_oil_change'] = dataset['sunflower_oil_change'].str.replace('%', '')
```

```python
dataset = dataset.convert_dtypes()
dataset.dtypes
```

```python
dataset = dataset.astype({
 'coconut_oil_price': float,
 'coconut_oil_change': float,
 'olive_oil_price': float,
 'olive_oil_change': float,
 'palmkernel_oil_price': float,
 'palmkernel_oil_change': float,
 'palm_oil_price': float,
 'palm_oil_change': float,
 'peanut_oil_price': float,
 'peanut_oil_change': float,
 'rapeseed_oil_price': float,
 'rapeseed_oil_change': float,
 'soybean_oil_price': float,
 'soybean_oil_change': float,
 'sunflower_oil_price': float,
 'sunflower_oil_change': float
})
```

```python
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(10)

sns.lineplot(data = dataset, x = 'month', y = 'coconut_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'olive_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'palmkernel_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'palm_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'peanut_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'rapeseed_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'soybean_oil_price', ax = ax, legend = 'full')
sns.lineplot(data = dataset, x = 'month', y = 'sunflower_oil_price', ax = ax, legend = 'full')
```
